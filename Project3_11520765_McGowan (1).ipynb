# Abstract 

This report presents a Monte Carlo simulation of thermal neutron transport through shielding materials of water, lead, and graphite. Random sampling techniques are used throughout: uniform random number generation is validated, an exponential free-path sampler was derived via inverse-transform sampling, and isotropic scattering directions are generated using spherical polar coordinates. A one-dimensional slab geometry was adopted in which neutrons enter perpendicularly at $x=0$ and undergo successive absorption or scattering events governed by macroscopic cross-sections that were computed from tabulated nuclear data. For each material and a range of slab thicknesses, the fractions of neutrons absorbed, reflected, and transmitted were recorded, and their statistical uncertainties quantified. Characteristic attenuation lengths are extracted by a weighted linear regression of the transmission data on a logarithmic scale. The attenuation lengths obtained for water, lead and graphite were $1.865 \pm 0.043$cm, $9.065 \pm 0.273$cm and $10.784 \pm 0.544$cm, respectively. The simulation correctly recovers the known $\approx$ 45 cm absorption mean free path of water and reveals the different shielding efficiencies of the three materials. As a stretch extension, the Woodcock method was applied to a two-material composite slab, with results cross-validated against the direct simulation. Furthermore, the convergence of the attenuation length $\lambda_\text{att}$ was investigated by computing its value for increasing numbers of simulated neutrons $N$ for each material. At small $N$ the estimates fluctuate significantly between runs due to large statistical noise, but as $N$ increases the values stabilise towards the true result, demonstrating convergence. The total execution time for the mandatory requirements of this project is on average less than 36 seconds, indicating proficient performance of this simulation.

# Introduction
This project will use Monte Carlo techniques to model the stochastic transport of thermal neutrons through shielding materials of varying thickness.
> **What is the Monte Carlo Method?** <br>
The Monte Carlo Method uses repeated random sampling to obtain numerical results for problems that are too complex to solve analytically. Rather than solving the neutron transport equations directly, we simulate the random behaviour of individual neutrons and collect statistics from a large number of trials.

The study of neutron interactions dates back to 1932 when James Chadwick discovered the neutron. Despite carrying no electric charge, neutrons interact strongly with atomic nuclei and penetrate matter much more deeply than charged particles. This makes neutrons suited for probing the internal structure of materials. Neutron scattering is now an essential technique for studying the structure and dynamics of condensed matter systems <a href="#ref1">[1]</a>. In this project, thermal neutrons are simulated passing through materials of different thicknesses to investigate transmission, reflection and absorption probabilities. By modelling neutron interactions at material boundaries, the attenuation properties of shielding materials can be quantified.

Understanding neutron transport has major practical applications. In nuclear reactors, moderators and shielding materials are used to control neutron populations and maintain safe operation. Similar neutron interaction principles are applied in radiation shielding and medical applications such as neutron-based radiotherapy, where understanding penetration depth and scattering behaviour is crucial for accurate dose delivery and patient safety.

## Generating Random Numbers
We will start off by confirming that random numbers can be generated using the Python in-built random number generator, and that the numbers produced are uniform over a chosen range.

For this, we will use `numpy.random.uniform(a, b, N)` which draws a sample of **N** random numbers from a uniform distribution in the range of [**a**,**b**) (including **a** but not including **b**).

```python
plt.close('all')
# Generate the sample from a uniform distribution over the half-open interval.
min_range = 5
max_range = 10
N = 100000
random_sample = np.random.uniform(min_range, max_range,N)

# Plot them in a histogram to see the distributions 
count, bins, _ = plt.hist(random_sample, bins = 50, density=True, edgecolor = 'black', linewidth = 0.4, color='#ff99cc')

# Uncertainty on density
bin_width = bins[1] - bins[0]
p = bin_width / (max_range - min_range)
sigma_density = np.sqrt(N * p * (1 - p)) / (N * bin_width)

#Plot error bars
bin_centres = 0.5 * (bins[:-1] + bins[1:])
plt.errorbar(bin_centres, count, yerr=sigma_density, fmt='o', markersize=1, color='black', ecolor='gray', elinewidth=1, capsize=2, label='Histogram with error bars')

# Plot uncertainty band
expected_density = 1 / (max_range - min_range)
plt.fill_between(bins, expected_density - sigma_density, expected_density + sigma_density, color='#e6e6ff', alpha=0.2, label='PDF uncertainty band')

# Expected uniform distribution
plt.plot(bins, np.ones_like(bins) * 0.2, linewidth=2, color='#9999ff')

#plt.hist(random_sample, bins=50, density=True, edgecolor='grey')
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Histogram of numpy.random.uniform() samples")

mean = np.mean(random_sample)
std = np.std(random_sample)
plt.legend()
plt.text(0.98, 0.05, f'Mean: {expected_density:.3f}\nStd Dev: {sigma_density:.3f}', transform=plt.gca().transAxes, fontsize=10, verticalalignment='bottom', horizontalalignment='right', bbox=dict(facecolor='white', alpha=0.8))

plt.show()
```

### Uniform Distribution Sampling — Investigating change in N
Now, we will use an interactive plot to demonstrate the convergence of a sampled uniform distribution to its theoretical probability density function (PDF) as the sample size $N$ increases.

A uniform distribution over $[a, b]$ has the PDF:

$$f(x) = \frac{1}{b - a}, \quad a \leq x \leq b$$

with expected density $f(x) = \frac{1}{10 - 5} = 0.2$ for this example.

### Uncertainty estimation

The per-bin density uncertainty is derived from binomial statistics. The probability that a single sample falls in any given bin of width $\Delta x$ is:

$$p = \frac{\Delta x}{b - a}$$

The uncertainty on the bin count is $\sqrt{Np(1-p)}$, giving a density uncertainty of:

$$\sigma_{\text{density}} = \frac{\sqrt{N p (1-p)}}{N \cdot \Delta x}$$

## Spectral Test

A spectral test is used to assess the quality of a random number generator (RNG) by checking for unwanted patterns or correlations in the generated numbers. A good RNG should produce numbers that are **uniformly distributed** and **uncorrelated**. This means that no visible structure should appear when plotted in multiple dimensions, like bands.

### How it Works

Points are generated in 3D using three consecutive random numbers as coordinates $(x, y, z)$. If the RNG is truly random, the points should appear as a **uniform scatter** with no visible patterns. If the RNG has poor spectral properties, the points will lie on a series of parallel hyperplanes, revealing hidden correlations.

### numpy.random.uniform()

`numpy.random.uniform()` uses the **Mersenne Twister** algorithm, which is a high-quality RNG with a period of $2^{19937} - 1$. When tested, the points should show no spectral issues — appearing as a uniform cloud in 3D space.

### randssp

`randssp.ipynb` is a multiplicative congruential uniform random number generator: $$x_{n+1} = (ax_n + c) \ mod \ m.$$ It is a poor-quality RNG that fails the spectral test. When its output is plotted in 3D, the points visibly align along parallel planes , demonstrating the kind of structural correlations that can introduce bias into Monte Carlo simulations.

```python
"""
RANDSSP Multiplicative congruential uniform random number generator.
Based on the parameters used by IBM's Scientific Subroutine Package.
The statement: r = randssp(m,n) generates an m-by-n random matrix.
The function can not accept any other starting seed.

This function uses the "bad" generator parameters that IBM used in several libraries in the 1960's.  
There is a strong serial correlation between three consecutive values.
"""
def randssp(p,q):

    # check whether x already exists. If not, initialise parameters for the linear congruential random number generator
    try: x
    except NameError:
        # modulus
        m = pow(2, 31)
        # multiplier
        a = pow(2, 16) + 3
        # increment
        c = 0
        # initial seed
        x = 123456789
    
    try: p
    except NameError:
        p = 1
    try: q
    except NameError:
        q = p

    # create empty array for random numbers
    r = np.zeros([p,q])

    # generate p × q pseudorandom values
    for l in range (0, q):
        for k in range (0, p):
            # x_(n+1) = (a*x_n + c) mod m
            x = np.mod(a*x + c, m)
            # normalise to range [0,1]
            r[k, l] = x/m
    
    return r
```

```python
# Create the RANDSSP
k = randssp(3, 10000)
# Create a good RNG with no spectral issues
random3d_sample = np.random.uniform(0,1,(3, 10000))
```

### Conclusion
The two plots clearly demonstrate the difference between a poor and a high-quality random number generator.

> Since the Marsaglias’ theorem states that all LCGs produce points that lie on hyperplanes, there are correlations between the points.
The spectral issues can be seen from the RANDSSP generator when looking along the z-axis.

The **RANDSSP** plot (left) shows points aligning along a series of distinct parallel planes in 3D space. This is a clear failure of the spectral test, as the points are not uniformly distributed but instead reveal hidden correlations between consecutive random numbers. This structure would introduce systematic bias into any Monte Carlo simulation, producing incorrect results.

The **numpy.random.uniform()** plot (right) shows points scattered uniformly throughout the unit cube with no visible structure or patterns. This confirms that `numpy.random.uniform()` passes the spectral test and is suitable for use in the neutron transport simulation.

This comparison highlights why the choice of RNG is critical in Monte Carlo methods; using RANDSSP in the neutron simulation would cause neutrons to preferentially travel along certain directions corresponding to the hyperplane structure, producing physically incorrect transmission, reflection and absorption rates.

#### The Von Neumann Rejection Method
The Von Neumann Rejection Method is a technique for generating random samples from a target distribution when direct sampling is difficult, using a simpler "proposal" distribution as a tool. It is not particularly necessary for this project, where an exponential function is used, but I have used it to demonstrate how sampling works when direct methods aren’t straightforward.

**How it works:**
1. Sample a candidate $X$ from $g(x)$
2. Sample $U$ ~ Uniform(0, 1), independently
3. Accept $X$ as a draw from $f$ if: $$U \leq \frac{f(X)}{c \cdot g(X)}$$
4. Reject otherwise and return to step 1

```python
# Exponential distribution using the Von Neumann Rejection Sampling Method
def exp_pdf(x, meanfreepath):
    return (1/meanfreepath) * np.exp(-x/meanfreepath)

# implementation of the Von Neumann Rejection Sampling 
def exp_sample_reject(meanfreepath, size):
    x_min = 0
    x_max = 5 * meanfreepath
    y_max = exp_pdf(0, meanfreepath) 
    
    samples_x = []
    samples_y = []
    
    while len(samples_x) < size:
        # sample x uniformly in [a, b]
        x_i = np.random.uniform(x_min, x_max)
        # evaluate PDF at x_i
        f_xi = exp_pdf(x_i, meanfreepath)
        # sample y uniformly in [0, y_max]
        y_i = np.random.uniform(0, y_max)

        # accept
        if y_i < f_xi:
            samples_x.append(x_i)
            samples_y.append(y_i)
    
    return np.array(samples_x), np.array(samples_y)
```

```python
# Fit data to an exponential fit
def fit_data(bins, error, std, mean_c, x):
    """
    """
    # Plot log as it will be linear for an exponential
    count_log = np.log(mean_c)
    # uncertainty on each bin
    weights = 1/error

    # fit data
    coeffs, cov = np.polyfit(bins, count_log, 1, w=weights, cov=True)
    grad = coeffs[0]
    intercept = coeffs[1]
    grad_err = np.sqrt(cov[0, 0])
    y = np.exp(intercept) * np.exp(grad * x)

    # residuals between log(counts) and fitted line
    y_bins  = grad * bins + intercept
    residuals   = count_log - y_bins

    # uncertainty on log(counts)
    sigma_log = 1 / np.sqrt(mean_c)
    chi2 = np.sum((residuals / sigma_log)**2)

    # degrees of freedom = number of bins - number of fit parameters
    dof = len(mean_c) - 2
    chi2_red = chi2 / dof
    
    return x, y, chi2_red, grad, grad_err, dof
```

### Stochastic Understanding

A single simulation run is subject to **statistical fluctuations** where the results vary slightly each time due to the random nature of the sampling. To obtain a reliable estimate of the true underlying distribution, the simulation is repeated $N_\text{iter}$ times, and the results are averaged:

$$\bar{f}(x) = \frac{1}{N_\text{iter}} \sum_{i=1}^{N_\text{iter}} f_i(x),$$

where $f_i(x)$ is the histogram from the $i$-th iteration. This gives a more stable estimate of the probability distribution and allows us to quantify the uncertainty across runs.

### Generating Exponential Step Lengths

Step lengths are drawn from the exponential distribution using the inverse CDF method:$$s_i = -\lambda \ln(u_i), \quad u_i \sim \text{Uniform}(0, 1),$$

where $\lambda$ is the mean free path of the material. Each iteration generates `num_particles` samples, which are binned into a histogram to approximate the probability density function $P(x) = \frac{1}{\lambda}e^{-x/\lambda}$.

Repeating this `num_iterations` times and averaging gives a robust stochastic estimate of the distribution with quantifiable uncertainty.

```python
# establish an empty array for the frequencies of the histogram
freq = np.zeros([num_iterations, num_bins])
    
# Iterate to make 10 histograms to have a stochastic understanding of the system
for i in range(num_iterations):
    freq[i,:], bin_edges = np.histogram(-mfp_water * np.log(np.random.uniform(size=num_particles)), bins = num_bins, range = (0, r_max))

# Get mean, standard deviation and error
mean_freq = np.mean(freq, axis = 0)
std_freq = np.std(freq, axis = 0)

# Remove the final bin edge as it will give one too many values
bin_width = np.diff(bin_edges)
bin_edges = bin_edges[:-1]
bins =  np.delete(bin_edges,np.argwhere(mean_freq == 0))

# Apply a mask to remove empty bins
bin_edges = np.delete(bin_edges,np.argwhere(mean_freq == 0))
std_freq = np.delete(std_freq,np.argwhere(mean_freq == 0))
mean_freq = np.delete(mean_freq,np.argwhere(mean_freq == 0))

# get error in frequency
err_freq = std_freq / mean_freq
```

```python
# mean free path
samples_x, samples_y = exp_sample_reject(mfp_water, size=num_particles)

x_arr = np.linspace(0, r_max, 300)
x_fit, y_fit, rchi2, slope, slope_err, dof = fit_data(bin_edges, err_freq, std_freq, mean_freq, x_arr)

#  Extract attenuation length and uncert 
# lambda = -1/slope
fitted_mfp = -1/slope
fitted_mfp_err = slope_err / slope**2  
```

## Producing Isotropic Unit Vectors
Using `isotropic_step_func` with `exponential=True`, a new function is created that generates 3D isotropic steps where the step length $r$ is drawn from an exponential distribution: $$P(r) = \frac{1}{\lambda} e^{-r/\lambda},$$

where $\lambda$ is the mean free path. The step length is sampled using the inverse CDF method: $$r = -\lambda \ln(u), \quad u \sim \text{Uniform}(0, 1)$$

The direction is chosen isotropically using spherical polar coordinates: $$x = r \sin\theta \cos\phi, \quad y = r \sin\theta \sin\phi, \quad z = r \cos\theta, $$

where $\cos\theta \sim \text{Uniform}(-1, 1)$ and $\phi \sim \text{Uniform}(0, 2\pi)$.

#### Producing The Function
For a spherical distribution, it can be seen below that the slices in the middle of a sphere have a larger radius than those at the top and the bottom of the sphere. Therefore, a bias is introduced due to the larger-radius slices having higher densities than those with smaller radii. We need to ensure that $\theta$ avoids the bias. 

```python
# Avoid the bias in theta by using the random distribution to produce more points at the centre
def isotropic_step_func(exponential, size, meanfreepath):
    theta = np.arccos(np.random.uniform(-1, 1, size))

    # Establish other parameters
    phi = np.random.uniform(0, 2 * np.pi, size)
    
    if exponential == True:
        # This gives the step lengths directly
        # rejection sampling is too slow in this case, so the cdf is used
        r = exp_cdf(meanfreepath, size)
    else:
        r = 1
    # Find x, y and z
    # x is transport direction
    z = r * np.cos(theta)
    x = r * np.sin(theta) * np.cos(phi)
    y = r * np.sin(theta) * np.sin(phi)
    
    return np.array([x, y, z])
```

## Visualising Neutron Random Walk
The plots below show the 3D random walks of 10 neutrons through each material for a slab of thickness $L = 10$ cm. Each neutron enters the slab perpendicularly at $x = 0$ in the $+x$ direction. The shaded region represents the slab boundaries at $x = 0$ and $x = L$, and the colour of each path indicates the outcome:

- **Indigo** → Absorbed
- **Cadetblue** → Reflected ($x < 0$)
- **Palevioletred** → Transmitted ($x > L$)

### Absorption Check

At each interaction point, a random number $u \sim \text{Uniform}(0, 1)$ is generated and compared to the absorption probability $P_a = \Sigma_a / \Sigma_T$:$$u < P_a \Rightarrow \text{absorbed}$$

Since $u$ is uniformly distributed between 0 and 1, the probability of this condition being true is exactly $P_a$ — meaning on average a fraction $P_a$ of neutrons are absorbed at each interaction, consistent with the branching ratio from the macroscopic cross sections.

If the neutron is not absorbed ($u \geq P_a$), it is scattered isotropically and the random walk continues.

```python
def simulate_neutron(L, material_index, track=False):
    """
    Simulates one neutron at a time through a slab, tracking its position and direction.

    Parameters:
        - L(float): length of slab in cm
        - material_index(int):  the index that determines what material the particle is in. 0 = Water, 1 = Lead, 2 = Graphite
        - track (bool): if True, records the full 3D position at every step for visualisation. If False, only the outcome is returned. 
                        Default = False.

    Returns:
        - outcome (str): the outcome of the neutron:
                         'transmitted' — exited through x > L_total
                         'reflected'   — exited through x < 0
                         'absorbed'    — absorbed within the slab
        - path (list or None): list of 3D position arrays [x, y, z] at each step if track=True, otherwise None.
    """
    # Establish parameters: mean free path and probability of absorption
    mfp      = additional_data[material_index][2] 
    prob_abs = additional_data[material_index][5]

    #Establish the beginning position and direction
    position  = np.array([0.0, 0.0, 0.0])
    # starts in +x
    direction = np.array([1.0, 0.0, 0.0]) 
    
    # if track is True, start path as a list containing the initial position [0,0,0], otherwise set it to None.
    # tracks the path of the neutron
    path = [position.copy()] if track else None

    # first step always in +x
    step_size = -mfp * np.log(np.random.uniform())
    position  = position + step_size * direction

    while True:
        # append the position to the path if trakcing is on
        if track:
            path.append(position.copy())

        # neutron is reflected if its position is outside of the slab (<0).
        if position[0] < 0:
            return "reflected", path
        # neutron is transmitted if its position is beyond the slab (>L).
        if position[0] > L:
            return "transmitted", path

        # neutron is absorbed a randomly generated number is less than the prob of absorption.
        if np.random.uniform() < prob_abs:
            return "absorbed", path

        # isotropic scatter: new direction
        # tracks full 3D paths
        cos_theta = np.random.uniform(-1, 1)
        sin_theta = np.sqrt(1 - cos_theta**2)
        direction = np.array([sin_theta * np.cos(np.random.uniform(0, 2 * np.pi)), sin_theta * np.sin(np.random.uniform(0, 2 * np.pi)), cos_theta])

        # increment position
        step_size = -mfp * np.log(np.random.uniform())
        position  += step_size * direction
```

```python
def run_simulation(L, material_index, N, num_runs=10):
    """
    Runs the simulation for all the particles using a vectorised stochastic approach.
    All N neutrons are simulated simultaneously using numpy arrays, with neutrons
    removed from the array as they are absorbed, reflected or transmitted.
    
    Parameters:
        - L (float): slab thickness in cm
        - material_index (int): the index that determines what material the particle is in. 0 = Water, 1 = Lead, 2 = Graphite
        - N (int): number of neutrons simulated per run
        - num_runs (int): the number of times the simulation is repeated and averaged over to reduce statistical fluctuations. Default = 10.

    Returns:
        - rates (dict): dictionary containing the fraction of neutrons that were transmitted, reflected and absorbed.
                        Keys: 'transmitted', 'reflected', 'absorbed'
        - uncs (dict): dictionary containing the statistical uncertainty on each rate, calculated using the binomial formula:
                       sigma = sqrt(p(1-p) / N_total)
                       Keys: 'transmitted', 'reflected', 'absorbed'
    """
    # Establish parameters: mean free path and probability of absorption
    mfp = additional_data[material_index][2]
    prob_abs = additional_data[material_index][5]
    
    num_ref = np.zeros(num_runs)
    num_tra = np.zeros(num_runs)
    num_abs = np.zeros(num_runs)

    for i in range(num_runs):
        x = np.zeros(N)
        # first step always in +x
        x = -mfp * np.log(np.random.uniform(size=N))
        
        while len(x) > 0:
            u_abs = np.random.uniform(size=len(x))
            
            # these have all scattered at least once - don't count as uncollided transmission
            num_ref[i] += np.count_nonzero(x < 0)
            num_tra[i] += np.count_nonzero(x > L)
            num_abs[i] += np.count_nonzero((x >= 0) & (x <= L) & (u_abs < prob_abs))
            
            # remove the above
            # could use 'x = np.delete(x,np.argwhere((x<0) | (x>L) | (u_abs < prob_abs)))' but will use a vectorised approach instead
            keep = ~((x < 0) | (x > L) | (u_abs < prob_abs))
            x = x[keep]
            
            cos_theta = np.random.uniform(-1, 1, size=len(x))
            sin_theta = np.sqrt(1 - cos_theta**2)
            # x += r * sin_theta * cos_phi
            # only the x- coordinate is needed
            x += -mfp * np.log(np.random.uniform(size=len(x))) * sin_theta * np.cos(np.random.uniform(0, 2 * np.pi, size=len(x)))

    N_total = N * num_runs
    # rates and uncertainties in a dictionary as it is faster than a list and does not rely on indexing
    rates = {"transmitted": np.mean(num_tra)/N,
             "reflected":   np.mean(num_ref)/N,
             "absorbed":    np.mean(num_abs)/N}
    uncs = {}
    for key in rates:
        uncs[key] = np.sqrt(rates[key] * (1 - rates[key]) / N_total)
    return rates, uncs
```

## Quantify The Scattering Process

```python
def fixed_thickness_summary(N, L=10.0):
    """
    Runs the neutron transport simulation at a fixed slab thickness and prints
    a summary table of the transmission, reflection and absorption rates for
    all three materials

    Parameters:
        - N (int): number of neutrons simulated per run
        - L (float): slab thickness in cm

    Returns:
    """
    material_names = ["Water", "Lead", "Graphite"]
    colours = {"transmitted": "palevioletred", "reflected": "cadetblue", "absorbed": "indigo"}
    outcomes = ["transmitted", "reflected", "absorbed"]
    labels = ["Transmitted", "Reflected", "Absorbed"]

    # create figure once before the loop
    fig = plt.figure(figsize=(18, 12))
    fig.suptitle(f"Neutron Scattering Summary (L = {L} cm, N = {N:,})", fontsize=14)

    for i, mat_name in enumerate(material_names):
        rates, error = run_simulation(L, i, N, num_runs=10)

        vals = [rates[o] for o in outcomes]
        errs = [error[o] for o in outcomes]  
        cols = [colours[o] for o in outcomes]

        # Table
        ax_table = fig.add_subplot(3, 3, i + 1)
        ax_table.axis("off")
        ax_table.set_title(mat_name, fontsize=12, fontweight="bold")
        table_data = [[f"{v:.4f} ± {e:.4f}"] for v, e in zip(vals, errs)]
        t = ax_table.table(cellText=table_data, rowLabels=labels, colLabels=["Rate"], loc="center", cellLoc="center")
        t.auto_set_font_size(False)
        t.set_fontsize(10)
        t.scale(1, 2)
        for (row, col), cell in t.get_celld().items():
            cell.set_edgecolor("grey")
            if row == 0:
                cell.set_facecolor("#4a4a8a")
                cell.set_text_props(color="white", fontweight="bold")
            elif row == 1:
                cell.set_facecolor("palevioletred")
            elif row == 2:
                cell.set_facecolor("cadetblue")
            elif row == 3:
                cell.set_facecolor("mediumpurple")

        # Pie chart 
        ax_pie = fig.add_subplot(3, 3, i + 4)
        ax_pie.pie(vals, labels=labels, colors=cols, autopct="%1.1f%%", startangle=90, textprops={"fontsize": 9})
        ax_pie.set_title(f"{mat_name} — proportions", fontsize=10)

        # Bar chart
        ax_bar = fig.add_subplot(3, 3, i + 7)
        bars = ax_bar.bar(labels, vals, color=cols, edgecolor="grey", linewidth=0.5)
        ax_bar.errorbar(labels, vals, yerr=errs, fmt="none", color="black", capsize=4, elinewidth=1)
        ax_bar.set_ylim(0, 1)
        ax_bar.set_ylabel("Fraction")
        ax_bar.set_title(f"{mat_name} — rates", fontsize=10)
        ax_bar.grid(axis="y", alpha=0.3)

    plt.tight_layout()
    plt.show()
```

## Determining Attenuation Lengths
The **mean free path** $\lambda_{mfp}$ is the average distance a neutron travels along its actual trajectory between successive interactions. The **attenuation length** $\lambda_{att}$, is the average distance projected along the *incident* direction — i.e. the depth into the material gained between interactions <a href="#ref2">[2]</a>. For a neutron travelling at angle $\theta$ to the incident direction: $$\lambda_{att} = \lambda_{mfp} \cos\theta$$

The two quantities are only equal when there is no angular deflection ($\theta = 0$).

#### Beer-Lambert Law

The Beer-Lambert law describes the attenuation of **uncollided** neutrons through a slab:

$$T = e^{-\alpha L},$$ 
where $\alpha = 1/\sigma_{att}$ is the attenuation coefficient and $L$ is the slab thickness<a href="#ref3">[3]</a>. This applies strictly to neutrons that traverse the slab without any interaction.

> In a Monte Carlo simulation, **total** transmission exceeds uncollided transmission because scattered neutrons can still drift forward and exit after one or more collisions. Counting these inflates the apparent $\lambda$ since the slab appears more transparent than it truly is.

To recover the theoretical $\lambda_{att} = 1/\alpha$, only neutrons whose first free path carries them beyond $L$ should be counted as transmitted. A neutron's first free path is sampled as:

$$d = -\lambda_{att}\ln(u), \quad u \sim \mathrm{Uniform}(0,1)$$

A neutron is uncollided-transmitted if $d > L$, i.e.:

$$-\lambda_{mfp}\ln(u) > L$$

The fraction of neutrons satisfying this is exactly $e^{-L/\lambda_{att}}$, recovering 
Beer-Lambert analytically.

Fitting $\ln(T)$ against $L$ yields a straight line with gradient $-1/\lambda_{att}$, from which the attenuation length is recovered as: $$\lambda_{att} = -\frac{1}{\text{gradient}}$$

For uncollided neutrons travelling normally to the slab ($\theta = 0$), the attenuation length equals the mean free path: $$\lambda_{att} = \lambda_{mfp} = \frac{1}{\sigma_{total}}$$

This equality holds only for the uncollided case. For total transmission (including scattered neutrons), $\lambda_{mfp} > \lambda_{att}$ since scattered neutrons travel longer path lengths to reach the same depth $L$.

```python
# initialise figure
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle("Attenuation length fits", fontsize=13)
attenuation_lengths =[]
attenuation_errors =[]

for ax, mat_name in zip(axes, material_names):
    res   = all_results[mat_name]
    Length_arr = np.array(res["L"])
    Transmitted_arr = np.array(res["transmitted"])
    error_arr = np.array(res["err_t"])

    # remove zero transmission points (log undefined)
    mask  = (Transmitted_arr > 0) 
    Length_final = Length_arr[mask]
    Transmitted_final = Transmitted_arr[mask]
    error_final = error_arr[mask]

    # natural log of transmitted
    ln_T  = np.log(Transmitted_final)
    # uncertainty on ln(T) by error propagation
    err_lnT = error_final / Transmitted_final 

    
    # weighted linear fit: ln(T) = m*L + c
    weights = Transmitted_final / error_final
    coeffs, cov = np.polyfit(Length_final, ln_T, 1, w=weights, cov=True)

    # attenuation length from -ve reciprocal gradient
    lam_att = -1.0 / coeffs[0]
    # error in attenuation length propagated and using the covariance matrix
    d_lam= lam_att**2 * np.sqrt(cov[0, 0])
    attenuation_lengths.append(lam_att)
    attenuation_errors.append(d_lam)

    # after, polyfit data
    # fitted ln(T) values
    y_fit = np.polyval(coeffs, Length_final) 
    # residuals = observed - expected
    residuals = ln_T - y_fit 
    # weighted by uncertainty
    chi2 = np.sum((residuals / err_lnT)**2) 
    # 2 parameters: slope and intercept
    dof = len(Length_final) - 2 
    chi2_red = chi2 / dof
    
    # plot
    ax.errorbar(Length_final, ln_T, yerr=err_lnT, fmt="o", color="cadetblue", ms=2, elinewidth=0.8, capsize=2, label="Data")
    x_axis = np.linspace(Length_final[0], Length_final[-1], 200) 
    ax.plot(x_axis, np.polyval(coeffs, x_axis), color="palevioletred", lw=1.5, label=f"λ = {lam_att:.2f} ± {d_lam:.2f} cm\nχ²_red = {chi2_red:.2f}")    
    ax.set_title(mat_name)
    ax.set_xlabel("L (cm)")
    ax.set_ylabel("ln(Transmission)")
    ax.legend(fontsize=8)
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()   
```

## Woodcock Method
#### What is the Woodcock Method?

The Woodcock method is a method used for simulating neutrons through two adjacent slabs of different materials. It eliminates the necessity to track the boundary between them. This method is good for heterogeneous systems with similar mean free paths and geometries with very regular arrangements, e.g. the geometry of a typical reactor.

#### The Problem it Solves

When a neutron crosses from one material to another, the mean free path changes discontinuously at the boundary. In standard sampling, this requires calculating the exact distance to the boundary and restarting the step, which involves complex geometry calculations.

#### How it Works

Instead of tracking boundaries, the Woodcock method uses a single **majorant cross section**:

$$\Sigma_T = \max(\Sigma_1, \Sigma_2)$$

This corresponds to the shortest mean free path across both materials:

$$\lambda_T = \frac{1}{\Sigma_T}$$

All neutrons take steps drawn from this single distribution regardless of which material they are in:

$$s_i = -\lambda_T \log(u_i)$$

To correct for the fact that steps are too short in the lower density material, each interaction is classified as either **real** or **fictitious**:

- **Real** (probability $\Sigma_\text{local}/\Sigma_T$): the neutron is either absorbed or scattered according to the local branching ratio
- **Fictitious** (probability $1 - \Sigma_\text{local}/\Sigma_T$): the neutron continues in the same direction as if nothing happened

```python
def woodcock_simulation(L, material_indices, N, num_runs=10):
    """
    Simulates neutron transport through multiple adjacent slabs of different materials using the Woodcock (delta-tracking) method.
    
    Instead of tracking material boundaries explicitly, a single majorant cross section Sigma_T = max(Sigma_1, Sigma_2, ...) is used
    everywhere, with fictitious interactions introduced in lower-density regions to correct for the over-sampled step lengths. 
    This avoids complex geometry calculations at material boundaries.

    Parameters:
        - L (float): thickness of each individual slab in cm
        - material_indices (list):  list of material indices defining the slab geometry 
                                    e.g. [1, 2] for Lead + Graphite
                                    0 = Water, 1 = Lead, 2 = Graphite
        - N (int): number of neutrons simulated per run
        - num_runs (int): number of times the simulation is repeated and averaged to reduce statistical fluctuations. Default = 10

    Returns:
        - rates (dict): fraction of neutrons transmitted, reflected and absorbed
                        Keys: 'transmitted', 'reflected', 'absorbed'
        - uncs (dict): statistical uncertainty on each rate using the binomial formula sigma = sqrt(p(1-p) / N_total)
                       Keys: 'transmitted', 'reflected', 'absorbed'
    """
    # extract material properties for each slab from additional_data
    # mean free paths, absorption probabilities and total cross sections
    mfps    = [additional_data[i][2] for i in material_indices]
    p_abs   = [additional_data[i][5] for i in material_indices]
    Sigmas  = [additional_data[i][1] for i in material_indices]

    # convert to numpy arrays for fast vectorised indexing
    mfps_arr   = np.array(mfps)
    p_abs_arr  = np.array(p_abs)
    Sigmas_arr = np.array(Sigmas)

    # majorant: smallest mfp and largest cross section across all slabs
    # all neutrons step using lam_maj regardless of which material they are in
    lam_maj = min(mfps)
    Sigma_T = max(Sigmas)

    # total slab thickness and number of slabs
    L_total = L * len(material_indices)
    n_slabs = len(material_indices)

    # counters for each outcome per run to perform a mean later
    num_ref = np.zeros(num_runs)
    num_tra = np.zeros(num_runs)
    num_abs = np.zeros(num_runs)

    for i in range(num_runs):
        # first step always in +x direction
        # all N neutrons start at x=0 and take their first step using the majorant mfp
        x = -lam_maj * np.log(np.random.uniform(size=N))

        # initialise angles
        # theta = pi/2 means sin(theta) = 1, cos(theta) = 0. So, the first step is entirely in the x direction.
        theta = np.full(N, np.pi/2)
        # phi = 0 means cos(phi) = 1, sin(phi) = 0
        # combined with theta = pi/2: x += r*sin(pi/2)*cos(0) = r*1*1 = r
        phi = np.full(N, 0.0)

        while len(x) > 0:
            # 1. count exits
            num_ref[i] += np.count_nonzero(x < 0)
            num_tra[i] += np.count_nonzero(x > L_total)

            # 2. remove reflected and transmitted
            inside = (x >= 0) & (x <= L_total)
            x = x[inside]
            theta = theta[inside]
            phi = phi[inside]

            # break if all neutrons have left
            if len(x) == 0:
                break

            # 3. which slab is each neutron in
            # divides each neutron's position by the slab thickness and rounds down to the slab index
            slab_index  = np.clip((x / L).astype(int), 0, n_slabs - 1)
            # macroscopic total cross section for the respective slab
            Sigma_local = Sigmas_arr[slab_index] 
            p_abs_local = p_abs_arr[slab_index]

            # 4. establish fictitious probability
            v = np.random.uniform(size=len(x))
            is_fic = v > Sigma_local / Sigma_T

            # 5. check if absorbed (real interactions only)
            u = np.random.uniform(size=len(x))
            absorbed = ~is_fic & (u < p_abs_local)
            num_abs[i] += np.count_nonzero(absorbed)

            # 6. remove absorbed
            keep = ~absorbed
            x = x[keep]
            theta = theta[keep]
            phi = phi[keep]
            is_fic = is_fic[keep]

            # 7. update angles for real scattered only
            real_scatter = ~is_fic
            new_cos_theta = np.random.uniform(-1, 1, size=np.count_nonzero(real_scatter))
            theta[real_scatter] = np.arccos(new_cos_theta)
            phi[real_scatter] = np.random.uniform(0, 2*np.pi, size=np.count_nonzero(real_scatter))

            # 8. step all surviving neutrons
            r = -lam_maj * np.log(np.random.uniform(size=len(x)))
            x += r * np.sin(theta) * np.cos(phi)

    N_total = N * num_runs
    rates = {"transmitted": np.mean(num_tra) / N,
             "reflected":   np.mean(num_ref) / N,
             "absorbed":    np.mean(num_abs) / N}
    uncs  = {key: np.sqrt(rates[key] * (1 - rates[key]) / N_total) for key in rates}
    return rates, uncs
```

```python
def simulate_neutron_woodcock(L, material_indices, track=False):
    """
    Simulates a single neutron travelling through multiple adjacent slabs of different materials using the Woodcock (delta-tracking) method.
    Used for visualising individual neutron random walks.

    At each step, the neutron travels a distance drawn from the majorant mean free path. The interaction is then classified as either 
    fictitious (no effect, direction unchanged) or real (absorbed or scattered) based on the local cross section at the neutron's position.

    Parameters:
        - L (float): thickness of each individual slab in cm
        - material_indices (list):  list of material indices defining the slab geometry 
                                    e.g. [1, 2] for Lead + Graphite
                                    0 = Water, 1 = Lead, 2 = Graphite
        - track (bool): if True, records the full 3D position at every step for visualisation. If False, only the outcome is returned. 
                        Default = False.

    Returns:
        - outcome (str): the outcome of the neutron:
                         'transmitted' — exited through x > L_total
                         'reflected'   — exited through x < 0
                         'absorbed'    — absorbed within the slab
        - path (list or None): list of 3D position arrays [x, y, z] at each step if track=True, otherwise None.
    """
    # extract material properties for each slab from additional_data
    # mean free paths, absorption probabilities and total cross sections
    mfps = [additional_data[i][2] for i in material_indices]
    p_abs = [additional_data[i][5] for i in material_indices]
    Sigmas = [additional_data[i][1] for i in material_indices]

    # convert to numpy arrays for fast indexing
    mfps_arr = np.array(mfps)
    p_abs_arr = np.array(p_abs)
    Sigmas_arr = np.array(Sigmas)

    # majorant: smallest mfp and largest cross section across all slabs
    # all neutrons step using lam_maj regardless of which material they are in
    lam_maj = min(mfps)
    Sigma_T = max(Sigmas)

    # total slab thickness and number of slabs
    L_total = L * len(material_indices)
    n_slabs = len(material_indices)

    # initialise position and angles
    position = np.array([0.0, 0.0, 0.0])
    theta = np.pi / 2   # perpendicular incidence
    phi = 0.0

    # track path
    path = [position.copy()] if track else None

    # first step in +x
    step_size = -lam_maj * np.log(np.clip(np.random.uniform(), 1e-10, 1))
    position[0] += step_size * np.sin(theta) * np.cos(phi)

    while True:
        # track path
        if track:
            path.append(position.copy())

        # check exits
        if position[0] < 0:
            return "reflected", path
        if position[0] > L_total:
            return "transmitted", path

        # determine slab - fast numpy indexing
        slab_idx = min(int(position[0] / L), n_slabs - 1)
        Sigma_local = Sigmas_arr[slab_idx]
        p_abs_local = p_abs_arr[slab_idx]

        # determine fictitious or real
        v = np.random.uniform()
        is_fic = v > Sigma_local / Sigma_T

        # check for real absorption
        if not is_fic:
            if np.random.uniform() < p_abs_local:
                return "absorbed", path

            # update angles for real scatter only
            # fictitious, theta and phi unchanged
            cos_theta = np.random.uniform(-1, 1)
            theta = np.arccos(cos_theta)
            phi = np.random.uniform(0, 2 * np.pi)

        # step using majorant mfp
        step_size = -lam_maj * np.log(np.clip(np.random.uniform(), 1e-10, 1))
        # adjust x,y and z position
        position[0] += step_size * np.sin(theta) * np.cos(phi)
        position[1] += step_size * np.sin(theta) * np.sin(phi)
        position[2] += step_size * np.cos(theta)
```

```python
def visualise_woodcock_comparison(L, n_neutrons, combos):
    """
    Visualises Woodcock (delta-tracking) neutron random walks across multiple slab geometry configurations in 3D.

    Produces a side-by-side set of 3D subplots — one per material combination — showing the trajectories of multiple
    neutrons simulated using the Woodcock method. Each neutron path is colour-coded by outcome (absorbed, reflected,
    or transmitted), and slab volumes are rendered as semi-transparent coloured planes for spatial reference.

    Parameters:
        - L (float): thickness of each individual slab in cm. 
        - n_neutrons (int): number of neutrons to simulate and plot per material combination.
        - combos (list of list of int): each inner list defines a slab geometry as a sequence of material indices, 
                                        e.g. [[0, 1], [1, 2]] renders two subplots:
                                        Water + Lead, and Lead + Graphite.
                                        0 = Water, 1 = Lead, 2 = Graphite.

    Returns:
        - None. Renders and displays a matplotlib figure via plt.show().
    """
    mat_labels  = ["Water", "Lead", "Graphite"]
    mat_colours = ["steelblue", "slategrey", "olivedrab"]   # one colour per material
    outcome_colours  = {"absorbed": "indigo", "reflected": "cadetblue", "transmitted": "palevioletred"}
    fictitious_colour = "lightgrey"

    n_plots = len(combos)
    fig = plt.figure(figsize=(7 * n_plots, 7))
    fig.suptitle(f"Woodcock random walks (L = {L} cm per slab, {n_neutrons} neutrons)", fontsize=13)

    for plot_idx, material_indices in enumerate(combos):
        L_total = L * len(material_indices)
        n_slabs = len(material_indices)
        title = " + ".join(mat_labels[i] for i in material_indices)

        ax = fig.add_subplot(1, n_plots, plot_idx + 1, projection="3d")
        ax.set_title(title, fontsize=10)

        # Slab volume planes 
        yy, zz = np.meshgrid(np.linspace(-8, 8, 2), np.linspace(-15, 15, 2))
        for s, mat_idx in enumerate(material_indices):
            x_left = s * L
            x_right = (s + 1) * L
            colour = mat_colours[mat_idx]

            # left and right boundary faces of this slab
            for x_face in (x_left, x_right):
                ax.plot_surface(np.full_like(yy, x_face), yy, zz, alpha=0.12, color=colour, linewidth=0, antialiased=False)

            # top and bottom faces (y = ±8) to close the slab volume
            xx_slab, zz_slab = np.meshgrid(np.linspace(x_left, x_right, 2), np.linspace(-15, 15, 2))
            for y_face in (-8, 8):
                ax.plot_surface(xx_slab, np.full_like(xx_slab, y_face), zz_slab, alpha=0.08, color=colour, linewidth=0, antialiased=False)

            # front and back faces (z = ±15)
            xx_slab2, yy_slab2 = np.meshgrid(np.linspace(x_left, x_right, 2), np.linspace(-8, 8, 2))
            for z_face in (-15, 15):
                ax.plot_surface(xx_slab2, yy_slab2, np.full_like(xx_slab2, z_face), alpha=0.08, color=colour, linewidth=0, antialiased=False)

            # material label
            ax.text((s + 0.5) * L, 9, 0, mat_labels[mat_idx], fontsize=8, ha="center", color="dimgrey")

        # Neutron tracks 
        for i in range(n_neutrons):
            # use simulate_neutron_woodcock with track=True to track its path
            outcome, path = simulate_neutron_woodcock(L, material_indices, track=True)
            path = np.array(path)
            col = outcome_colours.get(outcome, "black")
            ax.plot(path[:, 0], path[:, 1], path[:, 2], color=col, alpha=0.8, linewidth=0.9)

        # set elevation and axis labels
        ax.view_init(elev=30, azim=270)
        ax.set_xlabel("x (cm)")
        ax.set_ylabel("y (cm)")
        ax.set_zlabel("z (cm)")
        ax.set_xlim(-1, L_total + 1)

    # Legend 
    legend_elements = [Line2D([0], [0], color=c, label=l)
                       for l, c in outcome_colours.items()]
    legend_elements += [
        Patch(facecolor=mat_colours[i], alpha=0.4, label=mat_labels[i])
        for i in range(len(mat_labels))
    ]
    
    fig.legend(handles=legend_elements, loc="lower center", ncol=4, fontsize=9)
    plt.tight_layout(rect=[0, 0.05, 1, 1])
    plt.show()
```

```python
def woodcock_summary_table(L, N, num_runs=10):
    """
    Runs the woodcock simulation at a fixed slab thickness and prints a summary table of the transmission, reflection and 
    absorption rates for different combinations of materials.

    Parameters:
        - L (float): slab thickness in cm
        - N (int): number of neutrons simulated per run
        - num_runs (int): number of times the simulation is repeated and averaged to reduce statistical fluctuations. Default = 10

    Returns:
        - None. Renders and displays a matplotlib figure via plt.show().
    """
    combos = {
        "Lead + Graphite":          [1, 2],
        "Lead + Water":             [1, 0],
        "Lead + Graphite + Water":  [1, 2, 0],
    }
    colours  = {"transmitted": "palevioletred", "reflected": "cadetblue", "absorbed": "mediumpurple"}
    outcomes = ["transmitted", "reflected", "absorbed"]
    labels   = ["Transmitted", "Reflected", "Absorbed"]

    fig = plt.figure(figsize=(18, 12))
    fig.suptitle(f"Woodcock Simulation Summary (L = {L} cm per slab, N = {N:,})", fontsize=14)

    for plot_idx, (name, indices) in enumerate(combos.items()):
        rates, uncs = woodcock_simulation(L, indices, N, num_runs)
        vals = [rates[o] for o in outcomes]
        errs = [uncs[o]  for o in outcomes]
        cols = [colours[o] for o in outcomes]

        ax_table = fig.add_subplot(3, 3, plot_idx + 1)
        ax_table.axis("off")
        ax_table.set_title(name, fontsize=11, fontweight="bold")
        table_data = [[f"{v:.4f} ± {e:.4f}"] for v, e in zip(vals, errs)]
        t = ax_table.table(cellText=table_data, rowLabels=labels, colLabels=["Rate"], loc="center", cellLoc="center")
        t.auto_set_font_size(False)
        t.set_fontsize(10)
        t.scale(1, 2)
        for (row, col), cell in t.get_celld().items():
            cell.set_edgecolor("grey")
            if row == 0:
                cell.set_facecolor("#4a4a8a")
                cell.set_text_props(color="white", fontweight="bold")
            elif row == 1: cell.set_facecolor("palevioletred")
            elif row == 2: cell.set_facecolor("cadetblue")
            elif row == 3: cell.set_facecolor("mediumpurple")

        ax_pie = fig.add_subplot(3, 3, plot_idx + 4)
        ax_pie.pie(vals, labels=labels, colors=cols, autopct="%1.1f%%", startangle=90, textprops={"fontsize": 9})
        ax_pie.set_title(f"{name} — proportions", fontsize=9)

        ax_bar = fig.add_subplot(3, 3, plot_idx + 7)
        ax_bar.bar(labels, vals, color=cols, edgecolor="grey", linewidth=0.5)
        ax_bar.errorbar(labels, vals, yerr=errs, fmt="none", color="black", capsize=4, elinewidth=1)
        ax_bar.set_ylim(0, 1)
        ax_bar.set_ylabel("Fraction")
        ax_bar.set_title(f"{name} — rates", fontsize=9)
        ax_bar.grid(axis="y", alpha=0.3)

    plt.tight_layout()
    plt.show()
```

## Validating the Woodcock Method
The Woodcock method was validated by comparing the simulation results against the direct method by simulating two 5cm slabs of the same media side by side. In particular, the exponential attenuation of particles with depth was used as the primary benchmark, since the probability of survival in a medium is known to follow an exponential decay governed by the macroscopic cross-section.

To validate the method, simulated transmission, absorption, and reflection rates were computed for a range of slab thicknesses and compared against the expected exponential attenuation behaviour. Agreement between the simulated transmission curve and the theoretical exponential form indicates that the sampling of interaction lengths and collision events is physically consistent.

Further validation was carried out by checking that results were independent of the choice of Woodcock majorant cross-section. A correct implementation should produce identical macroscopic behaviour (within statistical uncertainty) even when the majorant is varied, provided it remains an upper bound on the total cross-section.

Finally, consistency was assessed by comparing results with an alternative approach (e.g. standard ray-tracing or direct sampling of mean free paths, if available). Close agreement between the two methods confirms that Woodcock sampling correctly reproduces particle histories without introducing bias.

Overall, the method is considered valid if it reproduces exponential attenuation, conserves expected probability balance (transmission + absorption + reflection ≈ 1), and agrees with benchmark simulations within statistical uncertainty.

```python
def woodcock_validation(N, num_runs=10):
    """
    Validates the Woodcock (delta-tracking) method against direct simulation by comparing neutron outcome rates for an equivalent
    geometry.

    Simulates neutrons through a Water + Water slab (2 × 5 cm) using the Woodcock method and through a single Water slab (10 cm)
    using direct simulation.
    Results are compared numerically and visually across two side-by-side panels:
        - Left: a styled table showing outcome fractions, uncertainties, and discrepancies in standard deviations.
        - Right: a grouped bar chart comparing Woodcock vs Direct outcome rates with error bars.

    Parameters:
        - N (int): number of neutrons simulated per run
        - num_runs (int): number of times the simulation is repeated and averaged to reduce statistical fluctuations. Default = 10

    Returns:
        - None. Renders and displays a matplotlib figure via plt.show().
    """
    L_val = 5.0
    outcomes = ["transmitted", "reflected", "absorbed"]
    labels = ["Transmitted", "Reflected", "Absorbed"]

    results_woodcock, errs_woodcock = woodcock_simulation(L_val, [0, 0], N, num_runs)
    results_direct, errs_direct = run_simulation(10.0, 0, N, num_runs)

    woodcock_vals = [results_woodcock[o] for o in outcomes]
    woodcock_errs = [errs_woodcock[o] for o in outcomes]
    direct_vals = [results_direct[o] for o in outcomes]
    direct_errs = [errs_direct[o] for o in outcomes]
    discrepancies   = [abs(w - d) / np.sqrt(we**2 + de**2)
               for w, d, we, de in zip(woodcock_vals, direct_vals, woodcock_errs, direct_errs)]

    fig2, (ax_t, ax_b) = plt.subplots(1, 2, figsize=(14, 5))
    fig2.suptitle("Woodcock Validation: Water+Water (5+5 cm) vs Direct Water (10 cm)", fontsize=13)

    ax_t.axis("off")
    ax_t.set_title("Numerical comparison", fontsize=11, fontweight="bold")
    table_data = [
        [f"{wv:.4f} ± {we:.4f}", f"{dv:.4f} ± {de:.4f}", f"{d:.2f}σ", "✓" if d < 2 else "✗"]
        for wv, we, dv, de, d in zip(woodcock_vals, woodcock_errs, direct_vals, direct_errs, discrepancies)
    ]
    t = ax_t.table(cellText=table_data, rowLabels=labels, colLabels=["Woodcock", "Direct", "Discrepancy", ""], loc="center", cellLoc="center")
    t.auto_set_font_size(False)
    t.set_fontsize(10)
    t.scale(1, 2.5)
    for (row, col), cell in t.get_celld().items():
        cell.set_edgecolor("grey")
        if row == 0:
            cell.set_facecolor("#4a4a8a")
            cell.set_text_props(color="white", fontweight="bold")
        elif row == 1: cell.set_facecolor("palevioletred")
        elif row == 2: cell.set_facecolor("cadetblue")
        elif row == 3: cell.set_facecolor("mediumpurple")

    x = np.arange(len(labels))
    width = 0.35
    ax_b.bar(x - width/2, woodcock_vals, width, label="Woodcock", color="#9999ff", edgecolor="grey", linewidth=0.5)
    ax_b.bar(x + width/2, direct_vals, width, label="Direct", color="#e6e6ff", edgecolor="grey", linewidth=0.5) 
    ax_b.errorbar(x - width/2, woodcock_vals, yerr=woodcock_errs, fmt="none", color="black", capsize=4, elinewidth=1)
    ax_b.errorbar(x + width/2, direct_vals, yerr=direct_errs, fmt="none", color="black", capsize=4, elinewidth=1)
    ax_b.set_xticks(x)
    ax_b.set_xticklabels(labels)
    ax_b.set_ylim(0, 1)
    ax_b.set_ylabel("Fraction")
    ax_b.set_title("Woodcock vs Direct — rates", fontsize=9)
    ax_b.legend(fontsize=10)
    ax_b.grid(axis="y", alpha=0.3)

    plt.tight_layout()
    plt.show()
```

```python
woodcock_validation(N=10000, num_runs=10)
```

## Showing convergence
>Plot how $\lambda_{att}$ changes as $N$ increases, demonstrating when the simulation has "converged".

As mentioned earlier, results in Monte Carlo simulations are subject to statistical fluctuations that decrease as the number of simulated particles $N$ increases. The simulation is said to have **converged** when increasing $N$ further produces no significant change in the result. Therefore, the estimate has stabilised to its true value.

The statistical uncertainty on any rate $p$ scales as: $$\sigma = \sqrt{\frac{p(1-p)}{N}}.$$

Hence, doubling $N$ reduces the uncertainty by a factor of $\sqrt{2}$. This means there are diminishing returns to increasing $N$ indefinitely.

***

The attenuation length $\lambda_\text{att}$ was computed for each material at a fixed slab thickness by running `run_simulation` with increasing values of $N$. The fitted $\lambda_\text{att}$ and its uncertainty $\sigma_\lambda$ are plotted against $N$ to show how the estimate stabilises.

```python
def transmission_model(L, lam, T0):
    return T0 * np.exp(-L / lam)

N_values = [100, 500, 1000, 5000, 10000, 50000]

# fixed thickness range for convergence test
thickness_ranges_conv = {
    "Water":    np.linspace(0.5, 15, 10),
    "Lead":     np.linspace(1,   60, 10),
    "Graphite": np.linspace(1,   60, 10),
}

convergence_results = {mat: {"lam": [], "err": []} for mat in material_names}

with Progress() as progress:
    total = len(N_values) * sum(len(thickness_ranges_conv[mat]) for mat in material_names)
    task  = progress.add_task("Convergence test...", total=total)

    for N in N_values:
        for i, mat_name in enumerate(material_names):
            thicknesses = thickness_ranges_conv[mat_name]
            trans, errs = [], []

            for L in thicknesses:
                r, u = run_simulation(L, i, N, num_runs=5)
                trans.append(r["transmitted"])
                errs.append(u["transmitted"])
                progress.advance(task)

            trans = np.array(trans)
            errs  = np.array(errs)
            mask  = trans > 0

            try:
                popt, pcov = curve_fit(transmission_model, thicknesses[mask], trans[mask], p0=[5.0, 1.0], sigma=errs[mask], absolute_sigma=True)
                lam, T0 = popt
                lam_err, _ = np.sqrt(np.diag(pcov))
            except RuntimeError:
                lam, lam_err = np.nan, np.nan

            convergence_results[mat_name]["lam"].append(lam)
            convergence_results[mat_name]["err"].append(lam_err)
```

```python
# Plot 
colours = {"Water": "cadetblue", "Lead": "palevioletred", "Graphite": "mediumpurple"}

fig, axes = plt.subplots(1, 3, figsize=(16, 5), sharey=False)
fig.suptitle("Convergence of attenuation length with N", fontsize=13)

for ax, mat_name in zip(axes, material_names):
    lams = np.array(convergence_results[mat_name]["lam"])
    errs = np.array(convergence_results[mat_name]["err"])
    col  = colours[mat_name]

    ax.errorbar(N_values, lams, yerr=errs, fmt="o-", color=col, capsize=4, elinewidth=1, linewidth=1.2, markersize=4, label="λ_att")

    # 1/sqrt(N) reference line scaled to the last point
    ref = lams[-1] + errs[-1] * np.sqrt(N_values[-1]) / np.sqrt(np.array(N_values))
    ax.plot(N_values, ref, "--", color="grey", linewidth=0.8, label="1/√N reference")

    # shade the final ± 1 sigma band as the "converged" value
    ax.axhline(lams[-1], color=col, linewidth=0.8, linestyle=":")
    ax.fill_between(N_values, lams[-1] - errs[-1], lams[-1] + errs[-1], alpha=0.15, color=col, label=f"Converged: {lams[-1]:.2f} ± {errs[-1]:.2f} cm")

    ax.set_xscale("log")
    ax.set_xlabel("N (neutrons)")
    ax.set_ylabel("λ_att (cm)")
    ax.set_title(mat_name)
    ax.legend(fontsize=8)
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

### Conclusion

At small $N$ (100) the estimates of $\lambda_\text{att}$ fluctuate significantly between runs due to large statistical noise. As $N$ increases towards 50000, the estimates converge toward a stable value and the uncertainty shrinks. The simulation can be considered converged once the error bars overlap consistently and the value no longer changes significantly with further increases in $N$.

This analysis helps identify the minimum $N$ needed to obtain a reliable result; balancing computational cost against statistical precision. Beyond 
the convergence point, additional neutrons provide little benefit while significantly increasing runtime.

# Final Discussion

A Monte Carlo simulation for thermal neutron transport was created and utilised on slabs of water, lead, and graphite across various thicknesses. At first, fundamental concepts were explored, such as confirming random number generation, building an exponential free-path sampler, and applying isotropic scattering. Subsequently, these elements were integrated into a complete neutron simulation within a one-dimensional slab.

The simulation accurately reflects the analytically anticipated attenuation length for water ($\approx$45 cm without scattering) and shows that the three materials display significantly different shielding properties. It was found that water possesses a large scattering cross-section, so it quickly absorbs neutrons, resulting in the shortest attenuation length among the three. This makes it the most efficient shield per unit mass for thermal neutrons. Lead offers moderate attenuation mainly due to its high density instead of a significant microscopic cross-section. Graphite has an insignificant absorption cross-section, allowing a relatively high proportion of neutrons to pass through even at considerable thicknesses. This is consistent with its role as a moderator rather than an absorber in reactor design. Attenuation lengths for each material were determined through weighted linear regression of ln(transmission rate) as a function of slab thickness, with propagation of their uncertainties based on counting statistics. The final attenuation lengths obtained for water, lead and graphite were $1.865 \pm 0.043$cm, $9.065 \pm 0.273$cm and $10.784 \pm 0.544$cm, respectively.

The Woodcock method was applied to a slab made of two materials and incorporated imaginary collisions to standardise the total cross-section throughout the domain. This removed the requirement to monitor specific material interfaces, while still converging to the same transmission fractions as direct simulations within statistical uncertainty. The method was tested on a two-material slab of the same material and gave transmission fractions consistent with the direct simulation within statistical uncertainty, confirming that both methods worked correctly.

Several advancements could be made to extend this project, such as:

- Testing **more materials** to give a broader picture of neutron attenuation properties

- Investigating **energy dependence** as this simulation used a single-group approximation where cross sections are fixed. A more realistic model would account for how cross sections vary with neutron energy as neutrons slow down

- Extending to **2D and 3D geometry** like cylindrical or spherical shielding geometries to be more representative of real reactor designs.

- **Increasing N** to reduce statistical uncertainties and give more precise attenuation lengths, as seen in the convergence investigation section.
